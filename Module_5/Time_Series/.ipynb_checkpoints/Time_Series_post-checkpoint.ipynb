## Seasonal Auto-Regressive Integrated Moving Averages (ARIMA)

##Let's talk about being stationary

Stationarity is one way of modeling the dependence structure between two variables.

### Constant Mean

![alt text]("Mean")


The mean of the series should not be a function of time. In the red graph, the mean increases over time. 

### Constant Variance

![alt text](http://www.analyticsvidhya.com/wp-content/uploads/2015/02/Var_nonstationary.png "Variance")


The variance of the series should not a be a function of time. This property is known as homoscedasticity. Notice the varying spread of distribution in the red graph.

### Autocovariance that does not depend on time.

![alt text](http://www.analyticsvidhya.com/wp-content/uploads/2015/02/Cov_nonstationary.png "Covariance")


The covariance of the i th term and the (i + m) th term should not be a function of time. In the following graph, you will notice the spread becomes closer as the time increases. Hence, the covariance is not constant with time for the ‘red series’.

### Why is stationarity important

It turns out that a lot of nice results which holds for independent random variables (law of large numbers, central limit theorem to name a few) hold for stationary random variables.

In the regression context, stationary  data allows us to model non-independent data.

### Two Steps to check for stationarity. 

**1. Plotting Rolling Statistics:** 
  * We can plot the moving average or moving variance and see if it varies with time. By moving average/variance I mean that at any instant ‘t’, we’ll take the average/variance of the last year, i.e. last 12 months.

**2. Dickey-Fuller Test:**
  * This is one of the statistical tests for checking stationarity. Here the null hypothesis is that the TS is non-stationary. If the ‘Test Statistic’ is less than the ‘Critical Value’, we can reject the null hypothesis and say that the series is stationary. 

## GMAT Example

```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime
import seaborn as sns
import statsmodels.api as sm  
from statsmodels.tsa.stattools import acf  
from statsmodels.tsa.stattools import pacf
from statsmodels.tsa.seasonal import seasonal_decompose
```

##Vizualize the time series

```python
df = pd.read_csv('Sales_Data.csv', sep='\t', index_col=0)
df.purchase_date = pd.to_datetime(df['purchase_date'])
df.purchase_date =  pd.DatetimeIndex(df['purchase_date']).normalize()
```

```python
daily_sales = df.groupby(['purchase_date'])['payment_amount'].sum()
daily_sales.head()
daily_sales.plot(figsize=(12,8), title= 'Daily Sales', fontsize=14)
```

###Aggregate the time period to months:

```python
monthly_sales = daily_sales.resample('m', how=sum)
monthly_sales.plot(figsize=(12,8), title=  'Monthly Sales', fontsize=14)
```

## Stationarize the Series

### How to make a time series stationary
1. Log Transform
2. Subtracting moving average
3. Differencing

```python
from statsmodels.tsa.stattools import adfuller
def test_stationarity(timeseries):
    
    #Determing rolling statistics
    rolmean = pd.rolling_mean(timeseries, window=12)
    rolstd = pd.rolling_std(timeseries, window=12)

    #Plot rolling statistics:
    fig = plt.figure(figsize=(12, 8))
    orig = plt.plot(timeseries, color='blue',label='Original')
    mean = plt.plot(rolmean, color='red', label='Rolling Mean')
    std = plt.plot(rolstd, color='black', label = 'Rolling Std')
    plt.legend(loc='best')
    plt.title('Rolling Mean & Standard Deviation')
    plt.show()
    
    #Perform Dickey-Fuller test:
    print 'Results of Dickey-Fuller Test:'
    dftest = adfuller(timeseries, autolag='AIC')
    dfoutput = pd.Series(dftest[0:4], index=['Test Statistic','p-value','#Lags Used','Number of Observations Used'])
    for key,value in dftest[4].items():
        dfoutput['Critical Value (%s)'%key] = value
    print dfoutput
```

```python
from statsmodels.tsa.stattools import adfuller
def test_stationarity1(timeseries):
    
    #Determing rolling statistics
    rolmean = pd.rolling_mean(timeseries, window=12)
    rolstd = pd.rolling_std(timeseries, window=12)

    #Plot rolling statistics:
    fig = plt.figure(figsize=(12, 8))
    orig = plt.plot(timeseries, color='blue',label='Original')
    mean = plt.plot(rolmean, color='red', label='Rolling Mean')
   # std = plt.plot(rolstd, color='black', label = 'Rolling Std')
    plt.legend(loc='best')
    plt.title('Rolling Mean & Standard Deviation')
    plt.show()
    
    #Perform Dickey-Fuller test:
    print 'Results of Dickey-Fuller Test:'
    dftest = adfuller(timeseries, autolag='AIC')
    dfoutput = pd.Series(dftest[0:4], index=['Test Statistic','p-value','#Lags Used','Number of Observations Used'])
    for key,value in dftest[4].items():
        dfoutput['Critical Value (%s)'%key] = value
    print dfoutput
```

```python
test_stationarity(monthly_sales)
```

```python
df_monthly_sales = monthly_sales.to_frame()
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose

decomposition = seasonal_decompose(df_monthly_sales['PAYMENT_AMOUNT'], freq=12)  
fig = plt.figure()  
fig = decomposition.plot()  
fig.set_size_inches(15, 8)
```

```python
df_monthly_sales['Natural_Log'] = df_monthly_sales.PAYMENT_AMOUNT.apply(lambda x: np.log(x))  
test_stationarity1(df_monthly_sales.Natural_Log)
```

```python
df_monthly_sales['First_Difference'] = df_monthly_sales.PAYMENT_AMOUNT - df_monthly_sales.PAYMENT_AMOUNT.shift(1)  
test_stationarity(df_monthly_sales.First_Difference.dropna(inplace=False))
```

```python
df_monthly_sales['First_Difference'] = df_monthly_sales.Natural_Log - df_monthly_sales.Natural_Log.shift(1)  
test_stationarity(df_monthly_sales.First_Difference.dropna(inplace=False))
```

```python
df_monthly_sales['seasonal_difference'] = df_monthly_sales.Natural_Log - df_monthly_sales.Natural_Log.shift(12)  
test_stationarity(df_monthly_sales.seasonal_difference.dropna(inplace=False))
```

```python
df_monthly_sales['difference_difference'] = df_monthly_sales.First_Difference - df_monthly_sales.First_Difference.shift(12)  
test_stationarity(df_monthly_sales.difference_difference.dropna(inplace=False))
```

## Finding optimal parameters


A seasonal ARIMA model is classified as an **ARIMA(p,d,q)x(P,D,Q)** model, 

    p=number of autoregressive (AR) terms 
    d=number of differences 
    q=number of moving average (MA) terms
   

    P=number of seasonal autoregressive (SAR) terms 
    D=number of seasonal differences 
    Q=number of seasonal moving average (SMA) terms

**Number of AR (Auto-Regressive) terms (p):** AR terms are just lags of dependent variable. 

**Number of MA (Moving Average) terms (q):** MA terms are the difference between the moving average at *i*th instant and actual value.

**Number of Differences (d):** In this case we took the first order difference. So either we can pass that variable and put d=0 or pass the original variable and put d=1. Both will generate same results.

## Plot ACF/PACF charts and find optimal parameters.


We use two plots to determine the (p,d,q) and (P,D,Q) numbers. Lets discuss the plots first.

**Autocorrelation Function (ACF):** It is a measure of the correlation between the the TS with a lagged version of itself. For instance at lag 5, ACF would compare series at time instant ‘t1’…’t2’ with series at instant ‘t1-5’…’t2-5’ (t1-5 and t2 being end points).

**Partial Autocorrelation Function (PACF):** This measures the correlation between the TS with a lagged version of itself but after eliminating the variations already explained by the intervening comparisons. Eg at lag 5, it will check the correlation but remove the effects already explained by lags 1 to 4.

```python
fig = plt.figure(figsize=(12,8))
ax1 = fig.add_subplot(211)
fig = sm.graphics.tsa.plot_acf(df_monthly_sales.First_Difference.iloc[1:], lags=40, ax=ax1)
ax2 = fig.add_subplot(212)
fig = sm.graphics.tsa.plot_pacf(df_monthly_sales.First_Difference.iloc[1:], lags=40, ax=ax2)
```

### Fit model to the data

```python
mod = sm.tsa.statespace.SARIMAX(df_monthly_sales['Natural_Log'], trend='n', order=(0,1,1), seasonal_order=(1,1,0,12))
results = mod.fit()
print results.summary()
```

### Predicting Sales
This code uses data up to July 2014 to predict forward the next 18 months

```python

df_monthly_sales['Forecast'] = results.predict(start = 56, end= 72, dynamic= True)  
df_monthly_sales[['Natural_Log', 'Forecast']].plot(figsize=(12, 8)) 
```

Our predictions were on the log transformed data, so we need to perform the inverse of this transformation to get back to real numbers

```python
df_monthly_sales['Delogged_forecast'] = df_monthly_sales.Forecast.apply(lambda x: np.exp(x))  
```

The chart below compares predictions starting in July 2014 through 2015. These predictions only use data upuntil July 2014 to make the predictions.

```python
npredict =df_monthly_sales.PAYMENT_AMOUNT['2015'].shape[0]
fig, ax = plt.subplots(figsize=(12,6))
npre = 12
ax.set(title='SALES', xlabel='Date', ylabel='dollars')
ax.plot(df_monthly_sales.index[-npredict-npre+1:], df_monthly_sales.ix[-npredict-npre+1:, 'PAYMENT_AMOUNT'], 'o', label='Observed')
ax.plot(df_monthly_sales.index[-npredict-npre+1:], df_monthly_sales.ix[-npredict-npre+1:, 'Delogged_forecast'], 'g', label='Dynamic forecast')
legend = ax.legend(loc='lower right')
legend.get_frame().set_facecolor('w')
```

While more test need ot be performed to get an standardize measure of the accuracy of these predictions, below is a brief measure.  Here we see the actual sales for 2015 compared the prediction made in July 2014.

```python
df_monthly_sales['2015'][['PAYMENT_AMOUNT', 'Delogged_forecast']].sum()
```

## Now let's predict all of 2016

```python
app = pd.DataFrame(index=pd.date_range('1/1/2016', periods=12, freq='m'), columns= df_monthly_sales.columns)
fin = pd.concat([df_monthly_sales, app])
```

```python
mod = sm.tsa.statespace.SARIMAX(fin.ix[:-12]['Natural_Log'], trend='n', order=(0,1,1), seasonal_order=(1,1,0,12))
results = mod.fit()
fin['Forecast'] = results.predict(start = 72, end= 84, dynamic= True)  
fin[['Natural_Log', 'Forecast']].plot(figsize=(12, 8)) 
```

```python
fin['Delogged_forecast'] = fin.Forecast.apply(lambda x: np.exp(x))  
```

```python
npredict =df_monthly_sales.PAYMENT_AMOUNT['2015'].shape[0]
fig, ax = plt.subplots(figsize=(12,6))
npre = 12
ax.set(title='SALES', xlabel='Date', ylabel='dollars')
ax.plot(fin.index[-npredict-npre+1:], fin.ix[-npredict-npre+1:, 'PAYMENT_AMOUNT'], 'b', label='Observed')
ax.plot(fin.index[-npredict-npre+1:], fin.ix[-npredict-npre+1:, 'Delogged_forecast'], 'g', label='Dynamic forecast')
legend = ax.legend(loc='lower right')
legend.get_frame().set_facecolor('w')
```
