##Let's talk about being stationary
You cannot build a time series model, unless your time series is stationary. In cases where the stationary criterion are violated, the first requisite becomes to stationarize the time series and then try stochastic models to predict this time series.

###Constant Mean

![alt text](http://www.analyticsvidhya.com/wp-content/uploads/2015/02/Mean_nonstationary.png "Mean")


The mean of the series should not be a function of time as seen in the red graph. 

###Constant Variance

![alt text](http://www.analyticsvidhya.com/wp-content/uploads/2015/02/Var_nonstationary.png "Variance")


The variance of the series should not a be a function of time. This property is known as homoscedasticity. Notice the varying spread of distribution in the red graph.

### Autocovariance that does not depend on time.

![alt text](http://www.analyticsvidhya.com/wp-content/uploads/2015/02/Cov_nonstationary.png "Covariance")


The covariance of the i th term and the (i + m) th term should not be a function of time. In the following graph, you will notice the spread becomes closer as the time increases. Hence, the covariance is not constant with time for the ‘red series’.

### Two Steps to check for staionarity. 

####1. Plotting Rolling Statistics: 
We can plot the moving average or moving variance and see if it varies with time. By moving average/variance I mean that at any instant ‘t’, we’ll take the average/variance of the last year, i.e. last 12 months.

####2. Dickey-Fuller Test:
This is one of the statistical tests for checking stationarity. Here the null hypothesis is that the TS is non-stationary. The test results comprise of a Test Statistic and some Critical Values for difference confidence levels. If the ‘Test Statistic’ is less than the ‘Critical Value’, we can reject the null hypothesis and say that the series is stationary. 

## How to make a time series stationary

###Eliminating the trend

1. Log Transform
2. Subtracting moving average

### Eliminating Seasonality

1. Differencing
2. Decomposition

Number of AR (Auto-Regressive) terms (p): AR terms are just lags of dependent variable. For instance if p is 5, the predictors for x(t) will be x(t-1)….x(t-5).

Number of MA (Moving Average) terms (q): MA terms are lagged forecast errors in prediction equation. For instance if q is 5, the predictors for x(t) will be e(t-1)….e(t-5) where e(i) is the difference between the moving average at ith instant and actual value.

Number of Differences (d): These are the number of nonseasonal differences, i.e. in this case we took the first order difference. So either we can pass that variable and put d=0 or pass the original variable and put d=1. Both will generate same results.

##How to determine the values of (p,d,q) and (P,D,Q)


We use two plots to determine these numbers. Lets discuss them first.

Autocorrelation Function (ACF): It is a measure of the correlation between the the TS with a lagged version of itself. For instance at lag 5, ACF would compare series at time instant ‘t1’…’t2’ with series at instant ‘t1-5’…’t2-5’ (t1-5 and t2 being end points).
Partial Autocorrelation Function (PACF): This measures the correlation between the TS with a lagged version of itself but after eliminating the variations already explained by the intervening comparisons. Eg at lag 5, it will check the correlation but remove the effects already explained by lags 1 to 4.

## GMAT Example

```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime
import seaborn as sns
import statsmodels.api as sm  
from statsmodels.tsa.stattools import acf  
from statsmodels.tsa.stattools import pacf
from statsmodels.tsa.seasonal import seasonal_decompose
```

##Vizualize the time series

```python
df = pd.read_csv('GMAT.csv', index_col=0)
df.DK_PURCHASE_DATE = pd.to_datetime(df['DK_PURCHASE_DATE'])
df.DK_PURCHASE_DATE =  pd.DatetimeIndex(df['DK_PURCHASE_DATE']).normalize()
df = df[df['PAYMENT_AMOUNT']>100]
daily_sales = df.groupby(['DK_PURCHASE_DATE'])['PAYMENT_AMOUNT'].sum()
daily_sales.plot(figsize=(12,8), title= 'GMAT Daily Sales', fontsize=14)
```

###Aggregate the time period to months:

```python
monthly_sales = daily_sales.resample('m', how=sum)
monthly_sales.plot(figsize=(12,8), title= 'GMAT Sales: Monthly Sales', fontsize=14)
```

##Stationarize the Series

```python
from statsmodels.tsa.stattools import adfuller
def test_stationarity(timeseries):
    
    #Determing rolling statistics
    rolmean = pd.rolling_mean(timeseries, window=12)
    rolstd = pd.rolling_std(timeseries, window=12)

    #Plot rolling statistics:
    orig = plt.plot(timeseries, color='blue',label='Original')
    mean = plt.plot(rolmean, color='red', label='Rolling Mean')
    std = plt.plot(rolstd, color='black', label = 'Rolling Std')
    plt.legend(loc='best')
    plt.title('Rolling Mean & Standard Deviation')
    plt.show()
    
    #Perform Dickey-Fuller test:
    print 'Results of Dickey-Fuller Test:'
    dftest = adfuller(timeseries, autolag='AIC')
    dfoutput = pd.Series(dftest[0:4], index=['Test Statistic','p-value','#Lags Used','Number of Observations Used'])
    for key,value in dftest[4].items():
        dfoutput['Critical Value (%s)'%key] = value
    print dfoutput
```

```python
test_stationarity(monthly_sales)
```

```python
df_monthly_sales = monthly_sales.to_frame()
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose

decomposition = seasonal_decompose(df_monthly_sales['Natural_Log'], freq=12)  
fig = plt.figure()  
fig = decomposition.plot()  
fig.set_size_inches(15, 8)
```

```python
df_monthly_sales['Natural_Log'] = df_monthly_sales.PAYMENT_AMOUNT.apply(lambda x: np.log(x))  
df_monthly_sales['Natural_Log'].plot(figsize=(12, 8)) 
```

```python

test_stationarity(df_monthly_sales.Natural_Log)
```

```python
df_monthly_sales['Logged_First_Difference'] = df_monthly_sales.Natural_Log - df_monthly_sales.Natural_Log.shift(1)  
df_monthly_sales['Logged_First_Difference'].plot(figsize=(12, 8)) 
```

```python
test_stationarity(df_monthly_sales.First_Difference.dropna(inplace=False))
```

```python
df_monthly_sales['seasonal_difference'] = df_monthly_sales.Natural_Log - df_monthly_sales.Natural_Log.shift(12)  
df_monthly_sales['seasonal_difference'].plot(figsize=(12, 8)) 
```

```python
test_stationarity(df_monthly_sales.seasonal_difference.dropna(inplace=False))
```

###Plot ACF/PACF charts and find optimal parameters.

```python
sm.graphics.tsa.plot_pacf(df_monthly_sales.seasonal_difference.iloc[12:])
sm.graphics.tsa.plot_acf(df_monthly_sales.seasonal_difference.iloc[12:])
```

### Fit model to the data

```python
mod = sm.tsa.statespace.SARIMAX(df_monthly_sales['Natural_Log'], trend='n', order=(0,0,0), seasonal_order=(2,2,0,12))
res = mod.fit()
print res.summary()
```

```python
results = mod.fit() 
df_monthly_sales['Forecast'] = results.predict(start = 54, end= 72, dynamic= True)  
df_monthly_sales[['Natural_Log', 'Forecast']].plot(figsize=(16, 12)) 
```

```python
df_monthly_sales['Delogged_forecast'] = df_monthly_sales.Forecast.apply(lambda x: np.exp(x))  
```
